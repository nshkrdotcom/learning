# Logit Lens

Reproduce the artifact from the repo root after a successful baseline run:

```bash
uv run python scripts/run_logit_lens.py --config configs/gpt2_small_induction.yaml --run runs/<run_id>
```

This notebook loads the generated CSV and figure.

````python
import json
from pathlib import Path

import pandas as pd
from IPython.display import Image, display

run_dir = Path("runs")
runs = sorted([p for p in run_dir.glob("*_gpt2_small_induction") if p.is_dir()])
latest = runs[-1] if runs else None
latest
````

````python
if latest and (latest / "logit_lens_by_layer.csv").exists():
    display(pd.read_csv(latest / "logit_lens_by_layer.csv").head())
    print(json.loads((latest / "logit_lens_summary.json").read_text()))
    figure = latest / "figures" / "logit_lens_expected_token.png"
    if figure.exists():
        display(Image(filename=str(figure)))
else:
    print("Missing logit-lens artifacts.")
````
