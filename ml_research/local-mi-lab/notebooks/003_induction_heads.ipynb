# Induction Heads

Reproduce the behavior, activation, and attention artifacts from the repo root:

```bash
uv run python scripts/run_baseline_behavior.py --config configs/gpt2_small_induction.yaml
uv run python scripts/cache_activations.py --config configs/gpt2_small_induction.yaml --run runs/<run_id>
uv run python scripts/run_attention_patterns.py --config configs/gpt2_small_induction.yaml --run runs/<run_id>
```

This notebook inspects tables and metadata. Attention-pattern evidence is descriptive, not causal evidence by itself.

````python
import json
from pathlib import Path

import pandas as pd
from IPython.display import Image, display

runs = sorted([p for p in Path("runs").glob("*_gpt2_small_induction") if p.is_dir()])
latest = runs[-1] if runs else None
latest
````

````python
if latest and (latest / "baseline_by_example.csv").exists():
    df = pd.read_csv(latest / "baseline_by_example.csv")
    display(df.sort_values("expected_rank").head())
    display(df.sort_values("expected_rank", ascending=False).head())
else:
    print("Missing baseline table.")
````

````python
manifest = latest / "activations" / "manifest.json" if latest else None
if manifest and manifest.exists():
    print(json.dumps(json.loads(manifest.read_text()), indent=2)[:4000])
else:
    print("Missing activation manifest.")
````

````python
attention_csv = latest / "attention_patterns_by_head.csv" if latest else None
if attention_csv and attention_csv.exists():
    attention = pd.read_csv(attention_csv)
    display(
        attention.sort_values("attention_to_previous_occurrence", ascending=False).head(20)
    )
else:
    print("Missing attention_patterns_by_head.csv.")
````

````python
attention_summary = latest / "attention_summary.json" if latest else None
if attention_summary and attention_summary.exists():
    print(json.dumps(json.loads(attention_summary.read_text()), indent=2)[:4000])
    figure = latest / "figures" / "attention_induction_scores.png"
    if figure.exists():
        display(Image(filename=str(figure)))
else:
    print("Missing attention summary artifacts.")
````
