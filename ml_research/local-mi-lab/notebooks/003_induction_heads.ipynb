# Induction Heads And Controls

Reproduce the controlled artifacts from the repo root:

```bash
uv run python scripts/build_toy_prompts.py --config configs/gpt2_small_induction_controls.yaml
uv run python scripts/run_baseline_behavior.py --config configs/gpt2_small_induction_controls.yaml
uv run python scripts/cache_activations.py --config configs/gpt2_small_induction_controls.yaml --run runs/<run_id>
uv run python scripts/run_logit_lens.py --config configs/gpt2_small_induction_controls.yaml --run runs/<run_id>
uv run python scripts/run_attention_patterns.py --config configs/gpt2_small_induction_controls.yaml --run runs/<run_id>
```

This notebook inspects generated tables and plots. Attention-pattern evidence is descriptive, not causal evidence by itself.

````python
import json
from pathlib import Path

import pandas as pd
from IPython.display import Image, display

controlled_runs = sorted([p for p in Path("runs").glob("*_gpt2_small_induction_controls") if p.is_dir()])
basic_runs = sorted([p for p in Path("runs").glob("*_gpt2_small_induction") if p.is_dir()])
latest = controlled_runs[-1] if controlled_runs else (basic_runs[-1] if basic_runs else None)
latest
````

````python
for name in ["baseline_by_example.csv", "baseline_by_family.csv"]:
    path = latest / name if latest else None
    if path and path.exists():
        print(name)
        display(pd.read_csv(path).head(20))
    else:
        print(f"Missing {name}.")
````

````python
manifest = latest / "activations" / "manifest.json" if latest else None
if manifest and manifest.exists():
    print(json.dumps(json.loads(manifest.read_text()), indent=2)[:4000])
else:
    print("Missing activation manifest.")
````

````python
for name in ["attention_patterns_by_head.csv", "attention_by_family.csv"]:
    path = latest / name if latest else None
    if path and path.exists():
        print(name)
        display(pd.read_csv(path).head(20))
    else:
        print(f"Missing {name}.")
````

````python
attention_summary = latest / "attention_summary.json" if latest else None
if attention_summary and attention_summary.exists():
    print(json.dumps(json.loads(attention_summary.read_text()), indent=2)[:5000])
    for figure_name in ["attention_induction_scores.png", "attention_by_family.png"]:
        figure = latest / "figures" / figure_name
        if figure.exists():
            display(Image(filename=str(figure)))
else:
    print("Missing attention summary artifacts.")
````

````python
path = latest / "logit_lens_by_family.csv" if latest else None
if path and path.exists():
    display(pd.read_csv(path).head(30))
else:
    print("Missing logit_lens_by_family.csv.")
````
